###`Q-1:` Create a random 3x4 matrix with value between 0-100. And perform below tasks

    i. Sort this matrix. np.sort()
    ii. Sort this matrix based on values in 2nd column.

    iv. Sort based on elements value.


````python
import numpy as np
a = np.array([[92, 90, 74],[ 6, 63, 93],[15, 93, 96],[70, 60, 48]])
print(a)
````

````python
np.sort(a)
````

````python
a[:,1].argsort()
````

````python
a[a[:,1].argsort()]
````

````python
a[np.max(a, axis=1).argsort()]
````

````python
a[np.max(a, axis=1).argsort()[::-1]]
````

### `Q-9:` You are given a array ($arr ∈ [0, 1]$). First you have round off the elements upto 3 decimal places and compare that
- 0th percentile == minimum value of the array
- 100th percentile == maximum value of the array
- also find the difference betwen 51th percenile and 50th percentile values

````python
# code here
arr = np.round(np.random.random((1,100)), 3)
arr
````

````python
if (np.percentile(arr, 0) == np.min(arr)):
  print("Yes", np.min(arr))
if (np.percentile(arr,100) == np.max(arr)):
  print("Yes", np.max(arr))
diff = np.percentile(arr, 51) - np.percentile(arr, 50)
print(np.percentile(arr, 51))
print(np.percentile(arr, 50))
print("Diff btw 51th and 50th percentile:", diff )
````
